```
### Importar tensorflow
import tensorflow as tf
```

```
### Importar pandas
import numpy as np
```

```
### Importar keras de tensorflow
from tensorflow import keras
```

```
 ### Definir numero de neuronas
 numNeuronas = 1
```

```
#### Creacion del modelo sequential
### el shape de entrada es de uno
model = keras.Sequential([keras.layers.Dense(units=numNeuronas, input_shape=[1])])
```

```
### Compliar el modelo definiiendo la metrica de perdida (loss) y el optimizador
model.compile(optimizer='sgd', loss='mean_squared_error', metrics='mean_squared_error')
```

```
### DEfinir los datos de entrada
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```
### Entrenar el modelo
### NumeroEpocas = 500
model.fit(xs, ys, epochs=500)
```

```
### Predecir
print(model.predict([10.0]))
```

```
### Path de la ubicacion del modelo
model_save_path = "/LRDeepLearning"
```

```
### Salvar el modelo
tf.saved_model.save(model, model_save_path)
```

```
### Prueba de carga del modelo
```

```
load_model = tf.saved_model.load(model_save_path)
```
